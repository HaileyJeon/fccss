```python
import urllib
from bs4 import BeautifulSoup
```

```python
url = "https://www.rottentomatoes.com"
html = urllib.request.urlopen(url)
source = html.read()
html.close()
print(source)
```

```python
soup = BeautifulSoup(source, "html.parser")
table = soup.find(id="Top-Box-Office")
print(table)
```

```python
all_tr = table.find_all("tr")
print(all_tr)
```

```python
for tr in all_tr:
    all_td = tr.find_all("td")
    score = all_td[0].find("span", attrs={"class":"tMeterScore"}).text
    link = url + all_td[1].a.get("href")
    mv_title = all_td[1].a.text
    amount = all_td[2].a.text
    amount = amount.split()[0]
    print(mv_title, "\n", score, amount, "\n",link, "\n")
```

```python
query = input("What do you want to find??")
url = "https://search.naver.com/search.naver?where=post&query=" + query

html = urllib.request.urlopen(url)
source = html.read()
html.close()
print(source)
```

```python
soup = BeautifulSoup(source, "html.parser")
ul = soup.find(id="elThumbnailResultArea")
all_li = ul.find_all("li")

for li in all_li:
    dt = li.find("dt")
    title = dt.a.get("title")
    link = dt.a.get("href")
    print(title, "\n", link)
```
